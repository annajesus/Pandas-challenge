# PyCity Schools Analysis

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

# District Summary

```python
# Total number of unique schools (ans:15)
school_count = school_data_complete['school_name'].unique()
school_count = len(school_count)
school_count
```

```python
# Total students (ans:39170)
student_count = school_data_complete['Student ID'].unique()
student_count = len(student_count)
student_count
```

```python
# Total budget (ans:24649428)
school_count_unique = school_data_complete['budget'].unique()
total_budget = school_count_unique.sum()
total_budget
```

```python
# Calculate the average math score (ans:78.98537145774827)
avg_math_score = school_data_complete['math_score'].mean()
avg_math_score
```

```python
# Calculate the average reading score (ans:81.87784018381414)
avg_reading_score = school_data_complete['reading_score'].mean()
avg_reading_score
```

```python
# % passing math (the percentage of students who passed math) (74.9808526933878)
passing_math_df = school_data_complete.loc[school_data_complete['math_score'] >= 70]
passing_math = len(passing_math_df)/ student_count * 100
passing_math
```

```python
# % passing reading (the percentage of students who passed reading) (85.80546336482001)
passing_reading_df = school_data_complete.loc[school_data_complete['reading_score'] >= 70]
passing_reading = len(passing_reading_df)/ student_count * 100
passing_reading
```

```python
# % overall passing (the percentage of students who passed math AND reading)(ans:65.17232575950983)
math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
math_reading_total = math_reading_count /  float(student_count) * 100
math_reading_total



```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame. 
district_key_metrics = pd.DataFrame({"Total Schools": school_count, "Total Students": student_count,
                                     "Total Budget": total_budget, "Avg Math Score": avg_math_score,
                                     "Avg Reading Score": avg_reading_score, "Percent Passing Math": passing_math,
                                     "Percent Passing Reading": passing_reading, "Overall Passing": math_reading_total}, index=[0])

#formatting
district_key_metrics["Total Budget"] = district_key_metrics["Total Budget"].map("${:,.0f}".format)
district_key_metrics["Percent Passing Math"] = district_key_metrics["Percent Passing Math"].map("{:,.2f}%".format)
district_key_metrics["Percent Passing Reading"] = district_key_metrics["Percent Passing Reading"].map("{:,.2f}%".format)
district_key_metrics["Overall Passing"] = district_key_metrics["Overall Passing"].map("{:,.2f}%".format)


# Display the DataFrame
district_key_metrics
```

# Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
range_labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df.head()
```

```python
# convert column Total student budget from string to float
school_spending_df['Total Student Budget'] = school_spending_df['Total Student Budget'].str.replace('$', '')
school_spending_df['Total Student Budget'] = school_spending_df['Total Student Budget'].astype(float)                                                                                            
                                                                                              
```

```python
# Use `pd.cut` to categorize spending based on the bins.                                                                                        
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['Total Student Budget'], spending_bins, labels=range_labels, right=False)
school_spending_df.head()
```

```python
# Groupby Spending ranges and calculate average
grouped_spending_df = school_spending_df.groupby(["Spending Ranges (Per Student)"])
school_spending_summary = grouped_spending_df.mean()
school_spending_summary.head()`
```

```python
school_spending_summary = school_spending_summary[["Avg Math Score", "Avg Reading Score", "Percent Passing Math", "Percent Passing Reading", "Percent Overall Passing"]]


# Display results
school_spending_summary
```
